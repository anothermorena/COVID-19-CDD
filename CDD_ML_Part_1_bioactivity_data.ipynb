Install the ChEMBL web service package so that we can retrieve bioactivity data from the ChEMBL Database.

## **Importing libraries**

```python
# Import necessary libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

## **Search for Target protein**
Target here reffers to the protein of interest or organism, for this project,I will be working with SARS-CoV2

### **Target search for coronavirus**

```python
# Target: search for coronavirus
target = new_client.target
target_query = target.search('coronavirus')
targets = pd.DataFrame.from_dict(target_query)
targets
```

### **Select and retrieve bioactivity data for *SARS Coronavirus 2 Replicase polyprotein 1ab* (seventh entry)**

We will assign the seventh entry (which corresponds to the target protein: *Replicase polyprotein 1ab*), to the ***selected_target*** variable 

```python
#Unique identification of the target
selected_target = targets.target_chembl_id[7]
selected_target
```

Here, we will retrieve only bioactivity data for *Replicase polyprotein 1ab* (CHEMBL4523582) that are reported as IC$_{50}$ values in nM (nanomolar) unit.

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
df.head(3)
```

lets see what other standard type values are present on the above df

```python
df.standard_type.unique()
```

For this dataset it would matter if we just went ahead without filtering as it only has the ideal unit of measurement, but for other datasets this might not be the case it could ec50 or %activity combined with ic50 and you might have to filter such a dataset. This is just to standardized our dataset to avoid having a mixture of bioactivity units

## **Handling missing data**
If any compounds has missing values for the **standard_value,canonical_smiles & molecule_chembl_id** columns, drop them. These are the features that will be extracted from this bioactivity data.

Standard value  is the potency of the drug, the lower the number the better and vice versa. So we want this value to be as low as possible.
Canonical smiles:  this is information about the chemical structure of that compound/molecule. A molecule or compund here means a drug . These are capable of producing a moduletory activity. In otherword, it excerts some effect (inhibit or activate) the target protein.
id: is self explanotory. Each compound is decribed by a chembl molecule ID.

```python
df2 = df[df.standard_value.notna()]
df2
```

Apparently, for this dataset there is no missing data for the standard value column,now lets drop missing values for the canonical smiles and id's columns if any.

```python
df3 = df2[df2.canonical_smiles.notna()]
df3.head(5)
```

```python
df4 = df3[df3.molecule_chembl_id.notna()]
df4.head(5)
```

## **Data pre-processing of the bioactivity data**

### **Labeling compounds as either being active, inactive**
The bioactivity data is in the IC50 unit. Compounds having values of less than 1000 nM will be considered to be **active** while those greater than 1000 nM will be considered to be **inactive**.

```python
bioactivity_class = []
for i in df4.standard_value:
  if float(i) >= 1000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
```

### **Iterate the *molecule_chembl_id* to a list**

```python
mol_cid = []
for i in df4.molecule_chembl_id:
  mol_cid.append(i)
```

### **Iterate *canonical_smiles* to a list**

```python
canonical_smiles = []
for i in df4.canonical_smiles:
  canonical_smiles.append(i)
```

### **Iterate *standard_value* to a list**

```python
standard_value = []
for i in df4.standard_value:
  standard_value.append(i)
```

### **Combine the 4 lists into a dataframe**

```python
data_tuples = list(zip(mol_cid, canonical_smiles, bioactivity_class, standard_value))
df5 = pd.DataFrame( data_tuples,  columns=['molecule_chembl_id', 'canonical_smiles', 'bioactivity_class', 'standard_value'])
```
