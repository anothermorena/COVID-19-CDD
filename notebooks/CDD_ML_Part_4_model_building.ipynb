### import required packages

```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

### load the dataset

```python
df = pd.read_csv('replicase_polyprotein_1ab_bioactivity_data_2class_pIC50_pubchem_fp.csv')
```

```python
df
```

###  Input features

replicase polyprotein 1ab dataset contains 881 input features and 1 output variable (pIC50 values)

```python
#drop the output variable
X = df.drop('pIC50', axis=1)
X
```

### Output features

```python
Y = df.pIC50
Y
```


### Data Split (80/20 ratio) 

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
X_train.shape, Y_train.shape
```

```python
X_test.shape, Y_test.shape
```

### Building a Regression Model using Random Forest

```python
model = RandomForestRegressor(n_estimators=100)
model.fit(X_train, Y_train)
r2 = model.score(X_test, Y_test)
r2
```

```python
Y_pred = model.predict(X_test)
Y_pred
```

### Model Performance

```python
print('Mean squared error (MSE): %.2f'
      % mean_squared_error(Y_test, Y_pred))
print('Coefficient of determination (R^2): %.2f'
      % r2_score(Y_test, Y_pred))
```

### Scatter Plot of Experimental vs Predicted pIC50 Values

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(color_codes=True)
sns.set_style("white")

ax = sns.regplot(Y_test, Y_pred, scatter_kws={'alpha':0.4})
ax.set_xlabel('Experimental pIC50', fontsize='large', fontweight='bold')
ax.set_ylabel('Predicted pIC50', fontsize='large', fontweight='bold')
ax.set_xlim(0, 12)
ax.set_ylim(0, 12)
ax.figure.set_size_inches(5, 5)
plt.show
```
